```python
from bs4 import BeautifulSoup
import requests
```

```python
#read URLs from a text file
def read_urls_from_file(file_path):
    with open(file_path, 'r') as file:
        urls = [line.strip() for line in file if line.strip()]
    return urls
file_path = r'C:\Users\96650\Downloads\urls.txt'
urls = read_urls_from_file(file_path)
```

```python

#lists to store elements
h2_elements = []
h3_elements = []
h4_elements = []
p_elements = []

content=[]

for url in urls:
 page = requests.get(url)
 soup = BeautifulSoup(page.text, "html.parser")
 h2_elements.extend(soup.find_all('h2'))
 h3_elements.extend(soup.find_all('h3'))
 h4_elements.extend(soup.find_all('h4'))
 p_elements.extend(soup.find_all('p'))
 
 headers=soup.find_all(['h2', 'h3', 'h4'])
 for header in headers:
    # Find the next paragraph sibling
    next_sibling = header.find_next_sibling()

    while next_sibling and next_sibling.name != 'p':
            next_sibling = next_sibling.find_next_sibling()

    if next_sibling and next_sibling.name == 'p':
            content.append({
                'header': header.text.strip(),
                'paragraph': next_sibling.text.strip()
            })
```

```python
for item in content:
    print(f"Header: {item['header']}")
    print(f"Paragraph: {item['paragraph']}")
    print('-' * 80)
```

```python
import csv

csv_file_path = r"C:\Users\96650\Downloads\GenAI.csv"

with open(csv_file_path, 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['Header', 'Paragraph']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    
    writer.writeheader()
    for item in content:
        writer.writerow({'Header': item['header'], 'Paragraph': item['paragraph']})

print(f"Data saved to {csv_file_path}")
```
